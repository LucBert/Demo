#subplots

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
years = list(range(2012, 2022))
ms_stock_prices = [25.4, 30, 40, 49, 59, 81.5, 100.6, 155.9, 220.9, 335.5]
apl_stock_prices = [20.4, 17.5, 25.7, 25.4, 27.4, 42.1, 52.5, 69.9, 127.5, 151.2]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(years, ms_stock_prices, label='Ms')
plt.plot(years, apl_stock_prices, label='Apple')
plt.xticks(years[::3])
plt.legend()
plt.show()
```

```python
fig_1 = plt.figure(figsize=(14, 4))
ax_1 = fig_1.add_subplot(1, 3, 1)
ax_2 = fig_1.add_subplot(1, 3, 2)
ax_3 = fig_1.add_subplot(1, 3, 3)
```

```python
fig_1 = plt.figure(figsize=(10,10))
ax_1 = fig_1.add_subplot(2, 2, 1)
ax_2 = fig_1.add_subplot(2, 2, 2)
ax_3 = fig_1.add_subplot(2, 2, 3)
ax_4 = fig_1.add_subplot(2, 2, 4)
ax_1.plot(years, ms_stock_prices, label='MS', color='red')
ax_4.plot(years, apl_stock_prices, label='Apple')
fig_1.legend()
plt.show()
```

```python
fig_1 = plt.figure(figsize=(10,10))
ax_1 = fig_1.add_subplot(2, 2, 1)
ax_2 = fig_1.add_subplot(2, 2, 2)
ax_3 = fig_1.add_subplot(2, 2, 3)
ax_4 = fig_1.add_subplot(2, 2, 4)
ax_1.plot(years, ms_stock_prices, label='MS', color='red')
ax_4.plot(years, apl_stock_prices, label='Apple')
ax_1.legend()
ax_4.legend()
plt.show()
```

```python
apple_stocks_series = pd.Series(apl_stock_prices)
```

```python
fig_1, axes = plt.subplots(2, 3, figsize=(10,6))
axes[0,1].plot(years, ms_stock_prices)
plt.show()
```

```python
fig_1, axes = plt.subplots(2, 3, figsize=(10,6))
apple_stocks_series.plot(ax=axes[0,1])
plt.show()
```
